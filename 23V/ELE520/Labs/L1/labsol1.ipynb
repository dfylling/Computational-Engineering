This illustrates the generation of points of computations in the variable x and furthermore the computations of the probability density function values in the vector p

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
from pdffuns import *
```

```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "last" # all | last | last_expr | none 
```

```python
def norm1D(my,Sgm,x):
    [n,d]=np.shape(x)
    p = np.zeros(np.shape(x))
    for i in np.arange(0, n):
        p[i] = 1 / (np.sqrt(2 * np.pi) * Sgm) * \
            np.exp(-1 / 2 * np.square((x[i] - my)) / (np.square(Sgm)))
    return p
```

```python
x=np.arange(0.25,0.35,0.005).reshape(-1,1)
my = 0.4
Sgm = 0.01
h = 100000
P = 1/25
p = np.log(h*P*norm1D(my,Sgm,x))

my = 0.2
Sgm = 0.01
h = 250
P = 24/25
p2 = np.log(h*P*norm1D(my,Sgm,x))
```

```python
fc=np.array([1,1,1])*0.4
fig, ax=plt.subplots(1,1)
fig.set_facecolor(fc)
ax.plot(x,x*0,'g.')
ax.plot(x,p,'b')
ax.plot(x,p2,'g')
ax.plot(x,p,'r.')
ax.grid(color='gray',linestyle='-.')
ax.tick_params(colors='yellow')
plt.show()
```

```python
x1 = np.arange(-10,10.5,0.5).reshape(-1,1)
x2 = np.arange(-9,10.5,0.5).reshape(-1,1)
my = np.array([1,1])
Sgm = np.array([[5,3],[3,5]])
p, x1v, x2v = norm2D(my, Sgm, x1, x2)
```
