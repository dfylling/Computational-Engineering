```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
from pdffuns import *
```

2D solution

```python
def baye(my, Sgm):
    x1 = np.arange(-10,10.5,0.5).reshape(-1,1)
    x2 = np.arange(-9,10.5,0.5).reshape(-1,1)

    p, x1v, x2v = norm2D(my, Sgm, x1, x2)
    fig = plt.figure(figsize =(9, 15))
    ax = plt.axes(projection ='3d')

    ax.plot_surface(x1v, x2v, p)
```

```python
my = np.array([3,3])
Sgm = np.array([[2,0],[0,2]]) 
baye(my, Sgm)
my = np.array([3,-2])
Sgm = np.array([[0.5,0],[0,2]]) 
baye(my, Sgm)
```
